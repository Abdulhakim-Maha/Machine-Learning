```python
def linear05(x):
	result = 0.5*x
	return result
```

```python
def linear1(x):
	result = x
	return result
```

```python
def linear2(x):
	result = 2*x
	return result
```

```python
linear05(3)
```

```python
linear1(3)
```

```python
linear2(3)
```

```python
import numpy as np
input = np.arange(-7,7,0.15)
output05 = [linear05(x) for x in input]
output1 = [linear1(x) for x in input]
output2 = [linear2(x) for x in input]
```

```python
import matplotlib.pyplot as plt

plt.plot(input, output05,lw = 2.0, c='g')
plt.plot(input, output1,lw = 2.0 ,c='m')
plt.plot(input, output2,lw = 2.0, c='b')
plt.ylabel('y (Output)')
plt.xlabel('x (Input)')
plt.axhline(y = 0.0, color = 'r', linestyle = '--', lw = 1.2)
plt.grid()
```
